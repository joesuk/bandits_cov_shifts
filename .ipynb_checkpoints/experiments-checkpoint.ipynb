```python
# visualize typical reward profile of synthetic environment

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

# parameters
K = 3 # number of arms
colors = ['b','r','g'] # colors to show dominance of each arm
cmaps = ['Blues','Reds','Greens'] # colormaps for each of the above colors
num_bumps = 25 # number of bumps in reward function
max_height = 0.3 # max reward difference over 0.5

# decide location, radii, heights, and signs of bumps for bump reward function
centers, radii, heights, signs, rads = construct_bumps(num_bumps, K, max_height,"gaussian")

res = 200 # resolution of space
xcoords = np.linspace(0,1,res)
ycoords = np.linspace(0,1,res)
X = np.array(np.meshgrid(xcoords,ycoords)).T.reshape(-1,2)
n = len(X[:,0])

f = reward_function(X, n, K, centers, radii, rads, heights, signs, num_bumps, max_height)

oracle = np.zeros(n,dtype=int)
gaps = np.zeros(n)
seconds = np.zeros(n)

for i in range(n):
    argmaxs = np.argwhere(f[i,:] == np.amax(f[i,:])).reshape(-1)
    optimalArm = np.random.choice(argmaxs)
    oracle[i] = optimalArm
    if len(argmaxs) == K:
        gaps[i] = 0
        seconds[i] = np.random.choice([i for i in argmaxs if i != optimalArm])
    else:
        second = 0
        for k in range(K):
            if (k not in argmaxs) and f[i,k] < f[i,optimalArm]:
                second = max(f[i,k],second)

        gaps[i] = f[i,optimalArm] - second
        seconds[i] = second


arm_points = [[] for i in range(K)]
arm_gaps = [[] for i in range(K)]

for i in range(0,n):
    argmaxs = np.argwhere(f[i,:] == np.amax(f[i,:])).reshape(-1)        
    for k in range(K):
        if k in argmaxs:
            arm_points[k].append([X[i,0],X[i,1]])
            arm_gaps[k].append(gaps[i])

fig, axes = plt.subplots(nrows=1,ncols=1)

for k in range(K):
    arm_gaps[k] = np.array(arm_gaps[k])
    arr = np.stack(arm_points[k],axis=0)
    axes.scatter(arr[:,0],arr[:,1], c=arm_gaps[k],cmap=cmaps[k])

axes.set_xlim((0,1))
axes.set_ylim((0,1))
axes.minorticks_off()
axes.set_xticks([], [])
axes.set_yticks([],[])

axes.set_xlabel(r"Average rewards over $[0, 1]^2$", fontsize=35,labelpad=20)

axes.set(adjustable='box', aspect='equal')

legend_elements = []
for k in range(K):
    legend_elements.append(mpatches.Patch(facecolor=colors[k],label=f'Arm {k+1}'))

leg = axes.legend(handles=legend_elements,loc='lower right',prop={'size': 20})
leg.set_title("Top Arm", prop = {'size':'xx-large'})


fig.tight_layout()
plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, 
            hspace = 0, wspace = 0)
fig.subplots_adjust(wspace=1)
fig.subplots_adjust(hspace=1)
plt.margins(0,0)
fig.tight_layout(pad=1)
fig.set_size_inches(6, 6)
plt.savefig("reward_profile.png", dpi=300,bbox_inches='tight', pad_inches=0.5)
```

```python
%autoreload 2

# experiments for varying source sample size and transfer exponent

## parameters

## range of transfer exponent values
gammas = np.logspace(np.log10(0.1), np.log10(20), num=20)
gammas = np.concatenate((gammas,np.logspace(np.log10(20), np.log10(40), num=20)))
gammas = np.insert(gammas,0,0)

## range of source sample size values
P_counts = np.trunc(np.logspace(np.log10(1), np.log10(10000), num=40)).astype(int)
P_counts = np.insert(P_counts,0,0).astype(int)

## other parameters
trials = 20 # number of trials
max_P = max(P_counts) 
n_Qg = 500 # target sample size for experiments varying gamma
n_Q = 500 # target sample size for experiments varying n_P
n_P = 500 # source sample size for experiments varying gamma
gamma = 1 # transfer exponent
n = max_P + n_Q # total horizon
K = 3 # number of arms
num_bumps = 25 # number of bumps
max_height = 0.3 # max height of reward differential at bumps

# construct Lipschitz environments
np.random.seed(1)
centers, radii, heights, signs, rads = construct_bumps(num_bumps, K, max_height,"gaussian")

# store regret vectors
cum_regrets_P = np.zeros((len(P_counts),trials))
cum_regrets_g = np.zeros((len(gammas),trials))

for j in range(trials):
    start_time = time.time()
    np.random.seed(j)

    X, Y, f, oracle = makeLipschitz(max_P, n_Q, K, gamma, centers, radii, rads, heights, signs, num_bumps, max_height)

    # experiments for changing source sample sizes
    for i in range(0,len(P_counts)):
        n_P = P_counts[i]
        pulls = adaptiveBandits(X, Y, n_P, n_Q, K, 1) 
        regret = computeRegret(pulls, n_P, n_Q, oracle, f, X, Y)
        cum_regrets_P[i,j] = regret[-1]

    # experiments for changing transfer exponents
    for g in range(0,len(gammas)):
        np.random.seed(int(j*g+100))
        X, Y, f, oracle = makeLipschitz(n_P, n_Qg, K, gammas[g], centers, radii, rads, heights, signs, num_bumps, max_height)
        start_time = time.time()
        pulls = adaptiveBandits(X, Y, n_P, n_Qg, K, 1) 
        regret = computeRegret(pulls, n_P, n_Qg, oracle, f, X, Y)
        cum_regrets_g[g,j] = regret[-1]

    np.savez('data/experiment.npz', P=cum_regrets_P, g=cum_regrets_g,trial=j)
    print("trial %d--- %s seconds ---" % (j,(time.time() - start_time)))
```
